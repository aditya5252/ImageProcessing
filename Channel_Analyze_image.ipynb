```python
import sys
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size': 22})

imBGR = cv2.imread('Data/FlowerBlur.jpg')
img = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)

print(f'The image dimensions are {img.shape}\n & \nThe image pixel dtpe is {img.dtype}')
```

```python
'Show image'
plt.figure(figsize=(8,8))
plt.imshow(img)
```

```python
'''Check each channel matrix values for a particular region '''
'''For Blue region'''
crop_img=img[880:1000,1300:1400,:]
plt.imshow(crop_img)
print(f'Here the R channel value is {crop_img[75,40,0]}')
print(f'Here the G channel value is {crop_img[75,40,1]}')
print(f'Here the B channel value is {crop_img[75,40,2]}')
print('Here Blue channel value dominates')
```

```python
'''Check each channel matrix values for a particular region '''
'''For Red region'''
crop_img=img[1450:1600,540:700,:]
plt.imshow(crop_img)
print(f'Here the R channel value is {crop_img[50,50,0]}')
print(f'Here the G channel value is {crop_img[50,50,1]}')
print(f'Here the B channel value is {crop_img[50,50,2]}')
print('Here Red channel value dominates')
```

```python
'''Check each channel matrix values for a particular region '''
'''For Green region'''
crop_img=img[1200:1400,540:700,:]
plt.imshow(crop_img)
print(f'Here the R channel value is {crop_img[50,50,0]}')
print(f'Here the G channel value is {crop_img[50,50,1]}')
print(f'Here the B channel value is {crop_img[50,50,2]}')
print('Here Green channel value dominates')
```

```python
import sys
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size': 22})

def downsample(img,factor):
    return img[::factor,::factor,:]

imBGR = cv2.imread('../Data/FlowerBlur.jpg')
img = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
```

```python
print(f'Initial image shape{img.shape}')
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.title('Original Image')
plt.show()

factor=100
down_img=downsample(img,factor)
print(f'Donwsampled image shape{down_img.shape}')
plt.figure(figsize=(8,8))
plt.imshow(down_img)
plt.title(f'Downsampled Image with factor {factor}')
plt.show()
```
